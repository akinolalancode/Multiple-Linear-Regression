# Looking at the intuition behind Gradient descent of Linear regression using Mean square error of 

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
X = np.array([1, 2, 3, 4, 5])  # Features
Y = np.array([2.2, 2.8, 4.5, 3.7, 5.5])  # Target values
m = len(X)  # Number of data points

# Hyperparameters
alpha = 0.01  # Learning rate
iterations = 1000  # Maximum number of iterations

# Initialize parameters
beta_0 = 0  # Intercept
beta_1 = 0  # Slope

# Keep track of MSE for each iteration
mse_history = []

# Gradient descent loop
for i in range(iterations):
    # Calculate predictions
    predictions = beta_0 + beta_1 * X
    
    # Compute Mean Squared Error
    mse = (1 / m) * np.sum((predictions - Y) ** 2)
    mse_history.append(mse)
    
    # Compute gradients
    db0 = (1 / m) * np.sum(predictions - Y)
    db1 = (1 / m) * np.sum((predictions - Y) * X)
    
    # Update parameters
    beta_0 -= alpha * db0
    beta_1 -= alpha * db1
    
    # Print progress at key iterations
    if (i) % 1 == 0 or i == 0:
        print(f"Iteration {i}:")
        print(f"  MSE = {mse:.4f}, beta_0 = {beta_0:.4f}, beta_1 = {beta_1:.4f}")
        print(f"  Predictions: {predictions}")
        print(f"  Actual:      {Y}")
        print("-" * 50)

    if (i) % 1 == 0 or i == 0:
        print(f"Iteration {i}:")
        print(f"  db0 = {db0:.4f},  db1 = {db1:.4f}")
        print(f"  db0: {db0}")
        print(f"  db1:      {db1}")
        print("-" * 50)
         
    if (i) % 1 == 0 or i == 0:
        print(f"Iteration {i}:")
        print(f"  beta_0 = {beta_0:.4f},  beta_1 = {beta_1:.4f}")
        print(f"  beta_0: {beta_0}")
        print(f"  beta_1:      {beta_1}")
        print("-" * 50)

# Final parameters
print(f"\nFinal Intercept (beta_0): {beta_0:.2f}")
print(f"Final Slope (beta_1): {beta_1:.2f}")

# Plot MSE over iterations
plt.figure(figsize=(8, 5))
plt.plot(range(1, iterations + 1), mse_history, color='blue')
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('MSE vs Iterations')
plt.show()

# Visualize the results
plt.scatter(X, Y, color="blue", label="Data points")
plt.plot(X, beta_0 + beta_1 * X, color="red", label="Regression line")
plt.xlabel("X (Feature)")
plt.ylabel("Y (Target)")
plt.title("Linear Regression using Gradient Descent")
plt.legend()
plt.show()
```
